```python
# Import the memory data from github
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from sklearn.cluster import KMeans
from scipy.spatial import distance
from numpy.linalg import matrix_rank

memory_data_url = "https://raw.githubusercontent.com/skhan102jh/Models-of-the-Neuron/main/MON_Memory_Variable.csv"
X = pd.read_csv(memory_data_url)
X = X.to_numpy()
```

# Question 1

```python
N = len(X)
W = (1/N)*np.matmul(X,X.transpose())
print("rank(W) = " +str(matrix_rank(W)))
```

$X$ is a $(1600 * 5)$ matrix, and $X^{T}$ is a $(5 * 1600)$ matrix, so $X X^{T}$ is a $(1600 * 1600)$ matrix. Therefore, $W = \frac{1}{N} X X^{T}$ is a $(1600 * 1600)$ matrix. 

$X$ contains the information of 5 independent memory images. Since each memory image is independent of the others, the five column vectors in $X$ are linearly independent. Thus, $rank(X) = 5$. We know from linear algebra that if we have a product of two matrices such that $Z = X Y$, then $rank(Z) \leq min(rank(X), rank(Y))$. As described in the lecture notes, $rank(W) = 1$. In this case, we get that $rank(X X^{T}) = 5$, so $rank(W) = 5$. 

# Question 2

By linear algebra we know that $(XY)^{T} =Y^{T} X^{T}$. We also know that if matrix $X$ is symmetric than $X^{T} = X$. Therefore, in order to prove that $W$ is symmetric, we must show that $W = W^{T}$. 

$W^{T} = \frac{1}{N} (X X^{T})^{T}$

$W^{T} = \frac{1}{N}(X^{T})^{T}(X)^{T}$

$W^{T} = \frac{1}{N} X X^{T}$

$W^{T} = W$

Therefore, $W$ is symmetric.


# Question 3

We will get the identity matrix because the diagonal terms will be representative of neurons within the same memory, so the states of the neuron within the same memory will be the same. I.e. that neuron will be $1$ or $-1$ in both instances of that memory. Therefore, the diagonal terms will be either $1*1 = 1$ or $-1*-1=1$. Thus, the diagonal terms will definitely be $1's$.

Additionally, the non diagonal terms will be representative of the same neuron across two different memories. The probability that that neuron will be in the same state (e.g. both in $1$ or both in $-1$) is $0.5$, resulting in $1$ being added to the sum. The probability that the neuron will be in different states (e.g. $1$ in one memory and $-1$ in the other memory) is also $0.5$, resulting in $-1$ being added to the sum. Therefore, the total sum of the $1's$ and $-1's$ will be approximately $0$. 

In summation, the diagonal terms of the matrix will definitely be $1's$ while the non diagonal terms will be approximately $0$, which is the form of an identity matrix.

```python
# Function to calculate du/dt
def u_prime(u, W):
  s = np.sign(u)
  return -1*u+np.matmul(W,s)


# Isolate the 5 separate images
mem_1 = np.reshape(X[:,0],(40,40))
mem_2 = np.reshape(X[:,1],(40,40))
mem_3 = np.reshape(X[:,2],(40,40))
mem_4 = np.reshape(X[:,3],(40,40))
mem_5 = np.reshape(X[:,4],(40,40))
#mem_5 = np.reshape(mem_5,(40,40))
mems = [mem_1,mem_2,mem_3,mem_4,mem_5]
mem_1_in = []; mem_2_in = []; mem_3_in = []; mem_4_in = []; mem_5_in = [];
mem_inputs = [mem_1_in,mem_2_in,mem_3_in,mem_4_in,mem_5_in]
# Create the upper left quadrant
for m in range(len(mems)): 
  #plt.figure()
  #plt.imshow(mems[m],cmap='gray')
  temp_input = np.reshape(np.zeros(1600),(40,40))
  for r in range(20):
    for c in range(20):
      temp_input[r][c] = mems[m][r][c]
  mem_inputs[m] = temp_input
  plt.figure()
  plt.imshow(mem_inputs[m],cmap='gray')
  plt.title("Memory Input " + str(m+1))
```

```python
# Method to recover image memory from input image
def recover_image(input):
  temp_input = np.reshape(input,(1600,1))
  time = np.arange(0, 1, .001)
  dt = 0.001 
  u = temp_input
  plt.figure()
  fig, ax = plt.subplots(ncols=2,figsize=(12,12))
  left = np.ravel(ax)[0].imshow(np.reshape(input,(40,40)).transpose(),cmap='gray')
  (ax)[0].set_title('Input Image')
  (ax)[0].set_xlabel('x pixels')
  (ax)[0].set_ylabel('y pixels')
  for t in range(1, len(time)):
    u = u + dt * u_prime(u,W)
  u = np.reshape(u,(40,40)).transpose()
  u = np.sign(u)
  #plt.imshow(u,cmap='gray')
  right = np.ravel(ax)[1].imshow(u,cmap='gray')
  (ax)[1].set_title('Recovered Image')
  (ax)[1].set_xlabel('x pixels')
  (ax)[1].set_ylabel('y pixels')
recover_image(mem_inputs[0])
recover_image(mem_inputs[1])
recover_image(mem_inputs[2])
recover_image(mem_inputs[3])
recover_image(mem_inputs[4])
```
